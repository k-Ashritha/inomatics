```python
# Smart Light Class

class SmartLight:
    def __init__(self, name):
        self.name = name
        self.status = "OFF"

    def turn_on(self):
        self.status = "ON"

    def turn_off(self):
        self.status = "OFF"

    def show_status(self):
        print(self.name, "is", self.status)


# Sample Input
light = SmartLight("Bedroom Light")
light.turn_on()

# Output
light.show_status()
```

```python
#problem 2
# Employee ID Card System

class Employee:
    def __init__(self, name, emp_id, department):
        self.name = name
        self.emp_id = emp_id
        self.department = department

    def display_id_card(self):
        print("Employee ID Card")
        print("Name:", self.name)
        print("ID:", self.emp_id)
        print("Department:", self.department)


# Sample Input
emp = Employee("Rahul", "EMP102", "IT")

# Output
emp.display_id_card()
```

```python
#problem 3
# Mobile Contact Record

class Contact:
    def __init__(self, name, phone):
        self.name = name
        self.phone = phone

    def display_contact(self):
        print("Contact Saved")
        print("Name:", self.name)
        print("Phone:", self.phone)


# Sample Input
c1 = Contact("Anita", "9876543210")

# Output
c1.display_contact()
```

```python
#problem 4
# Product Price Tag Generator

class Product:
    def __init__(self, name, price):
        self.name = name
        self.price = price

    def print_tag(self):
        print("Product:", self.name)
        print("Price: ₹" + str(self.price))


# Sample Input
p1 = Product("Headphones", 2499)

# Output
p1.print_tag()
```

```python
#problem 5
# Movie Rating Display System

class Movie:
    def __init__(self, name, rating):
        self.name = name
        self.rating = rating

    def display(self):
        print("Movie:", self.name)
        print("Rating:", self.rating, "/ 5")


# Sample Input
m1 = Movie("Inception", 4.8)

# Output
m1.display()
```

```python
#problem 6
# Delivery Address Manager

class Delivery:
    def __init__(self, customer, address):
        self.customer = customer
        self.address = address

    def show_details(self):
        print("Delivery Details")
        print("Customer:", self.customer)
        print("Address:", self.address)


# Sample Input
d1 = Delivery("Suman", "Hyderabad")

# Output
d1.show_details()
```
